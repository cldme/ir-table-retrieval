```python
# Util function to convert list of strings to one string   
def listToString(s):  
    
    # initialize an empty string
    str1=""
    
    if(len(s) > 0):
        str1 = s[0] 
    
    # traverse in the string   
    for ele in s[1:]:  
        str1 += " " + ele   
    
    # return string   
    return str1     
```

```python
## create list of all the unique tokens for the tables
import re
table_tokens=[]
cnt = 0
for t in data:
    cnt+=1
    #print(t)
    #for k,v in data[t].items():
    #   print ()
    tokens = listToString(data[t]["title"]);
    for d in data[t]["data"]:
        tokens = tokens + " " +listToString(d)
    tokens = tokens + " " + data[t]["caption"]
    table_tokens.append(set(re.split('; |, | |_|\*|\n|\]|\[',tokens)))
print(cnt)    
```
